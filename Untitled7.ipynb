```python
import datetime as dt
def get_year_day(timestamp):
    return dt.datetime.utcfromtimestamp(timestamp).strftime("%j")


df['YEARDAY'] =  df['TIMESTAMP'].apply(get_year_day)

#print(df['TIMESTAMP'])


is_the_same_yearday = 'true' if df['YEARDAY'].all() == df['YEARDAY'].any() else 'false'
print(is_the_same_yearday)

def get_month_day(timestamp):
    return dt.datetime.utcfromtimestamp(timestamp).strftime("%d")


df['MONTHDAY'] =  df['TIMESTAMP'].apply(get_month_day)




is_the_same_monthday = 'true' if df['MONTHDAY'].all() == df['MONTHDAY'].any() else 'false'
print(is_the_same_monthday)



def get_week_day(timestamp):
    return dt.datetime.utcfromtimestamp(timestamp).strftime("%w")


df['WEEKDAY'] =  df['TIMESTAMP'].apply(get_week_day)


is_the_same_weekday = 'true' if df['WEEKDAY'].all() == df['WEEKDAY'].any() else 'false'
print(is_the_same_weekday)

def convert_timestamp(timestamp):
    return int(dt.datetime.utcfromtimestamp(timestamp).strftime("%H"))*60.0 + int(dt.datetime.utcfromtimestamp(timestamp).strftime("%m"))

df['TIMESTAMP'] = df['TIMESTAMP'].apply(convert_timestamp)



def find_nearest(value):  
    array = np.array([int(18*60), int(8*60+30), int(4*60), int(14*60+30)])
    idx = (np.abs(array-(int(value)))).argmin()
    return idx

df['TIMESTAMP'] = df['TIMESTAMP'].apply(find_nearest)

```

```python


def total_time(POLYLINE):
    return float(len(POLYLINE[1:-1].replace("],[", "]_[").split("_"))*15)

df['TOTAL_TIME'] = df['POLYLINE'].apply(total_time)
```

```python
def add_first_lon(POLYLINE):
    return float(POLYLINE[1:-1].replace("],[", "]_[").split("_")[0][1:-1].split(",")[0])

df['FIRST_LON'] = df['POLYLINE'].apply(add_first_lon)

def add_first_lat(POLYLINE):
      return float(POLYLINE[1:-1].replace("],[", "]_[").split("_")[0][1:-1].split(",")[1])

df['FIRST_LAT'] = df['POLYLINE'].apply(add_first_lat)
```

```python
from random import randint

def get_cut_lon(POLYLINE):
    arr = POLYLINE[1:-1].replace("],[", "]_[").split("_")
    lenght = len(arr)
    cut = randint(0, lenght-1)
    return float(arr[cut][1:-1].split(",")[0])
  

def get_cut_lat(POLYLINE):
    arr = POLYLINE[1:-1].replace("],[", "]_[").split("_")
    lenght = len(arr)
    cut = randint(0, lenght-1)
    return float(arr[cut][1:-1].split(",")[1])
    

df['CUT_LAST_LON'] = df['POLYLINE'].apply(get_cut_lon)

df['CUT_LAST_LAT'] = df['POLYLINE'].apply(get_cut_lat)

df['DIF_LON']  = df['CUT_LAST_LON'] - df['FIRST_LON']
df['DIF_LAT']  = df['CUT_LAST_LAT'] - df['FIRST_LAT']
```

```python

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.cross_validation import ShuffleSplit, KFold


def rmsle(y_true, y_pred):
    return np.sqrt(np.mean((np.log(y_true + 1) - np.log(y_pred + 1))**2))
    
def accurancy(y_true, y_pred):
    return sum(np.absolute(x-y) for x,y in zip(y_true,y_pred)) / sum(y_true)

    
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

def splitDataset(dataset, dataset_y, splitRatio):
    # Training set size
    trainSize = int(dataset.shape[0] * splitRatio)
    
    # List of randomly chosen indicies
    indices = np.random.permutation(dataset.shape[0])
    
    # Split indicies for training and test set by trainSize
    training_idx, test_idx = indices[:trainSize], indices[trainSize:]
    
    print("training_idx", training_idx)
    print("test_idx", test_idx)
    # Create training and test sets by indicies
    training = dataset.iloc[list(training_idx)]
    test = dataset.iloc[list(test_idx)]
    
    training_y = dataset_y.iloc[list(training_idx)]
    test_y = dataset_y.iloc[list(test_idx)]
    
    return training, training_y, test, test_y

train_x_a0, train_y_a0, test_x_a0, test_y_a0 = splitDataset(df_A0_x_train, df_A0_y_train, 0.99)
    
# Initialize the famous Random Forest Regressor from scikit-learn
clf_RA0 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RA0.fit(train_x_a0, train_y_a0)
y_pred_rf_a0 = clf_RA0.predict(test_x_a0)

train_x_a1, train_y_a1, test_x_a1, test_y_a1 = splitDataset(df_A1_x_train, df_A1_y_train, 0.99)

# Initialize the famous Random Forest Regressor from scikit-learn
clf_RA1 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RA1.fit(train_x_a1, train_y_a1)
y_pred_rf_a1 = clf_RA1.predict(test_x_a1)


train_x_a2, train_y_a2, test_x_a2, test_y_a2 = splitDataset(df_A2_x_train, df_A2_y_train, 0.99)

# Initialize the famous Random Forest Regressor from scikit-learn
clf_RA2 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RA2.fit(train_x_a2, train_y_a2)
y_pred_rf_a2 = clf_RA2.predict(test_x_a2)


train_x_a3, train_y_a3, test_x_a3, test_y_a3 = splitDataset(df_A3_x_train, df_A3_y_train, 0.99)


# Initialize the famous Random Forest Regressor from scikit-learn
clf_RA3 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RA3.fit(train_x_a3, train_y_a3)
y_pred_rf_a3 = clf_RA3.predict(test_x_a3)

 
train_x_b0, train_y_b0, test_x_b0, test_y_b0 = splitDataset(df_B0_x_train, df_B0_y_train, 0.99)

    
# Initialize the famous Random Forest Regressor from scikit-learn
clf_RB0 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RB0.fit(train_x_b0, train_y_b0)
y_pred_rf_b0 = clf_RB0.predict(test_x_b0)

train_x_b1, train_y_b1, test_x_b1, test_y_b1 = splitDataset(df_B1_x_train, df_B1_y_train, 0.99)


# Initialize the famous Random Forest Regressor from scikit-learn
clf_RB1 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RB1.fit(train_x_b1, train_y_b1)
y_pred_rf_b1 = clf_RB1.predict(test_x_b1)


train_x_b2, train_y_b2, test_x_b2, test_y_b2 = splitDataset(df_B2_x_train, df_B2_y_train, 0.99)


# Initialize the famous Random Forest Regressor from scikit-learn
clf_RB2 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RB2.fit(train_x_b2, train_y_b2)
y_pred_rf_b2 = clf_RB2.predict(test_x_b2)

train_x_b3, train_y_b3, test_x_b3, test_y_b3 = splitDataset(df_B3_x_train, df_B3_y_train, 0.99)

# Initialize the famous Random Forest Regressor from scikit-learn
clf_RB3 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RB3.fit(train_x_b3, train_y_b3)
y_pred_rf_b3 = clf_RB3.predict(test_x_b3)

train_x_c0, train_y_c0, test_x_c0, test_y_c0 = splitDataset(df_C0_x_train, df_C0_y_train, 0.99)

    
# Initialize the famous Random Forest Regressor from scikit-learn
clf_RC0 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RC0.fit(train_x_c0, train_y_c0)
y_pred_rf_c0 = clf_RC0.predict(test_x_c0)

train_x_c1, train_y_c1, test_x_c1, test_y_c1 = splitDataset(df_C1_x_train, df_C1_y_train, 0.99)


# Initialize the famous Random Forest Regressor from scikit-learn
clf_RC1 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RC1.fit(train_x_c1, train_y_c1)
y_pred_rf_c1 = clf_RC1.predict(test_x_c1)

train_x_c2, train_y_c2, test_x_c2, test_y_c2 = splitDataset(df_C2_x_train, df_C2_y_train, 0.99)


# Initialize the famous Random Forest Regressor from scikit-learn
clf_RC2 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RC2.fit(train_x_c2, train_y_c2)
y_pred_rf_c2 = clf_RC2.predict(test_x_c2)

train_x_c3, train_y_c3, test_x_c3, test_y_c3 = splitDataset(df_C3_x_train, df_C3_y_train, 0.99)


# Initialize the famous Random Forest Regressor from scikit-learn
clf_RC3 = RandomForestRegressor(n_estimators=50, n_jobs=4, random_state=23)
clf_RC3.fit(train_x_c3, train_y_c3)
y_pred_rf_c3 = clf_RC3.predict(test_x_c3)


    
# or the Gradient Boosting Regressor
#clf = GradientBoostingRegressor(n_estimators=200, max_depth=3, random_state=23)
#clf.fit(X_trn, y_trn)
#y_pred_gb[val_idx] = clf.predict(X_tst)
        
#print('  Score RFR/GBR: %.4f, %.4f' % (rmse(df_A0_y_train, y_pred_rf)))
print('  Score RFR/GBR AO: %.4f' % (rmsle(test_y_a0, y_pred_rf_a0)))
print('  Score RFR/GBR A1: %.4f' % (rmsle(test_y_a1, y_pred_rf_a1)))
print('  Score RFR/GBR A2: %.4f' % (rmsle(test_y_a2, y_pred_rf_a2)))
print('  Score RFR/GBR A3: %.4f' % (rmsle(test_y_a3, y_pred_rf_a3)))

print('  Score RFR/GBR BO: %.4f' % (rmsle(test_y_b0, y_pred_rf_b0)))
print('  Score RFR/GBR B1: %.4f' % (rmsle(test_y_b1, y_pred_rf_b1)))
print('  Score RFR/GBR B2: %.4f' % (rmsle(test_y_b2, y_pred_rf_b2)))
print('  Score RFR/GBR B3: %.4f' % (rmsle(test_y_b3, y_pred_rf_b3)))


print('  Score RFR/GBR CO: %.4f' % (rmsle(test_y_c0, y_pred_rf_c0)))
print('  Score RFR/GBR C1: %.4f' % (rmsle(test_y_c1, y_pred_rf_c1)))
print('  Score RFR/GBR C2: %.4f' % (rmsle(test_y_c2, y_pred_rf_c2)))
print('  Score RFR/GBR C3: %.4f' % (rmsle(test_y_c3, y_pred_rf_c3)))

print('  Avarage: %.4f' % ((rmsle(test_y_a0, y_pred_rf_a0)+rmsle(test_y_a1, y_pred_rf_a1)+rmsle(test_y_a2, y_pred_rf_a2)+rmsle(test_y_a3, y_pred_rf_a3)+rmsle(test_y_b0, y_pred_rf_b0)+rmsle(test_y_b1, y_pred_rf_b1)+rmsle(test_y_b2, y_pred_rf_b2)+rmsle(test_y_b3, y_pred_rf_b3)+rmsle(test_y_c0, y_pred_rf_c0)+rmsle(test_y_c1, y_pred_rf_c1)+rmsle(test_y_c2, y_pred_rf_c2)+rmsle(test_y_c3, y_pred_rf_c3))/12))

y = [1, 2, 3, 4]

```
