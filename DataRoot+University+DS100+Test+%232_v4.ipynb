#### 7.1. Знайдіть найближче значення в матриці B до будь-якого довільно заданого скаляра.
```
```
```
```
```
```
```
Максимальна оцінка: 2 бал
Рекомендований час: 2хв
```

````python
m = np.array([[1, 24, 1, 8, 15], 
              [-3, 5, 7, -14, 16], 
              [4, 6, -3, 20, 2], 
              [10, 12, 19, 21, 3], 
              [1, 1, 2, -2, 0]])
def fun(x):
    closest = m[0,0]
    dif = np.absolute(closest - x)
    for i in m:
        temp = np.absolute(m[i] - x)
        if(temp < dif):
            dif = temp
            closest = m[i]
    return closest
    
print(fun(-5))
````
